```python
import pandas as pd
import numpy as np

from bs4 import BeautifulSoup
```

```python
def parse_table(table, len_table):
    result = np.empty([0,2])
    for i in range(2,len_table):
        a = str(table[i])
        delim = a.find('title=')
        if delim == -1:
            continue
        a = a[delim + 6:]
        delim = a.find('>')
        a = a[delim:]
        delim = a.find('<')
        album = a[1:delim]
        album = album.replace("'", "")
        delim = a.find('US:')
        if delim == -1:
            delim = a.find('US</a>:')
            if delim == -1:
                continue
        a = a[delim:]
        for i, c in enumerate(a):
            if c.isdigit():
                delim = i
                break
        a = a[delim:]
        delim = a.find('<')
        sale = a[:delim]
        sale = float(sale.replace(',',''))
        result = np.append(result, [[album, sale]],axis = 0)
    return result
```

```python
with open ("wiki_json.txt", "r") as my_file:
    data = my_file.readlines()
```

```python
with open ("artists.txt", "r") as my_file:
    artists = my_file.readlines()
```

```python
len_file = len(data)
```

```python
wiki_result = np.empty([0,3])
for i in range(len_file):
    html_doc = data[i]
    soup = BeautifulSoup(html_doc, 'html.parser')
    tr = soup.find_all('tr')
    length = len(tr)
    result = parse_table(tr, length)
    length = len(result)
    artist = [[artists[i] for j in range(1)] for j in range(length)]
    tempo = np.hstack((artist, result))
    wiki_result = np.append(wiki_result, tempo, axis = 0)
```

```python
col_names = ['Artist', 'Album', 'US sales']
df = pd.DataFrame(data=wiki_result, columns=col_names)
```

```python
df['Artist'] = df['Artist'].str.replace('\n','')
df
```

```python
df.to_csv('data_1.csv', index=False)
```
